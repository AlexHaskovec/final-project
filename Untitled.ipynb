```python
import pandas as pd

crimes = pd.read_csv("crime.csv", index_col="INCIDENTNEIGHBORHOOD")
census = pd.read_csv("population-density.csv")

crimes=crimes.drop(["CCR","INCIDENTTIME","CLEAREDFLAG","INCIDENTZONE","INCIDENTHIERARCHYDESC","OFFENSES","COUNCIL_DISTRICT","PUBLIC_WORKS_DIVISION"], axis=1)
print(len(crimes))

row=0
while row<len(crimes):
    if crimes.iloc[row].HIERARCHY >8:
        crimes.iloc[row].HIERARCHY=1
        row=row+1
    else:
        crimes.iloc[row].HIERARCHY=2
        row=row+1

crimes = crimes.groupby("INCIDENTNEIGHBORHOOD").sum()
crimes
```

```python
census = pd.read_csv("population-density.csv",index_col="Neighborhood")
census = census[['Pop. 2010']]

cenus=census.groupby("Neighborhood").sum()


mtolicrimes=(crimes.loc["Mount Oliver"].HIERARCHY+crimes.loc["Mt. Oliver Neighborhood"].HIERARCHY+crimes.loc["Mt. Oliver Boro"].HIERARCHY)



drop_crimes=crimes.drop(['Central North Side', 'Golden Triangle/Civic Arena',"Mount Oliver", "Mt. Oliver Neighborhood", "Mt. Oliver Boro", 'Outside City',
                 'Outside County', 'Outside State', 'Troy Hill-Herrs Island'])

mtoli = pd.DataFrame([["Mt. Oliver",mtolicrimes]], columns=(["INCIDENTNEIGHBORHOOD","HIERARCHY"]))
mtoli=mtoli.groupby("INCIDENTNEIGHBORHOOD").sum()


drop_crimes=drop_crimes.append(mtoli)

drop_crimes=drop_crimes.sort_index()



neighborhoodlist = []
for val in drop_crimes.index:
    neighborhoodlist.append(val)


myList=[]
i=0
while i<len(drop_crimes):
    myList.append(drop_crimes.iloc[i].HIERARCHY/census["Pop. 2010"][i])
    i+=1

data = {'Neighborhood':neighborhoodlist, 'Crimes_Per_Person':myList}
dataFrame = pd.DataFrame(data)

dataFrame.groupby("Crimes_Per_Person").sum()
```
